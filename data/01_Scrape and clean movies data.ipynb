# 2. Make a list of urls that you would want to interact with

1. The reason why we can make a keywords dict here is that, by looking at the urls, we can find a pattern
2. The pattern here is: https://movie.douban.com/people/53218548/collect?start={???}&sort=time&rating=all&filter=all&mode=grid, {???} is the only difference between each url.
3. That way, we can make a dict and put different numbers {???} into it, so that we can loop each url in a list later.

```python
# For a GET request, all of the necessary info is in the URL
collects = [
    "0",
    "15",
    "30",
    "45",
    "60",
    "75",
]

urls=[]

for collect in collects:
    url = f"https://movie.douban.com/people/53218548/collect?start={collect}&sort=time&rating=all&filter=all&mode=grid"
    #print("I want to scrape", url)
    urls.append(url)
```

# So, what you need to do is to make another empty list and append them

2. Make an empty list --> loop in your list of url --> append the doc(or whatever you want to doc.select/find_all(...)) 

3. Then accessing the data from your new list ( here my new list called: results )

P.S.: My cheatsheet: https://stackoverflow.com/questions/70191038/how-to-loop-through-list-of-url-and-download-specific-data-from-them-in-python-b

## The right way to do is shown below:

```python
# empty list to store all results
results = []

# your loop here
for url in urls:
    response = requests.get(url, cookies=cookies, headers=headers)
    doc = BeautifulSoup(response.text)
    results.append(doc.select(".item"))

# Accessing the data from the results
for result in results:
    print(result)
```

# 4. This is the real loop step! Watch out!

* Loop here
* and make a dataframe

```python
myfilm=[]

for result in results:
    for movie in result:
        for movie in movie.find_all("ul"):
            for title in movie.select(".title"):
                for url in title.find_all("a"):
                    title = url.text.strip()
                    url = url["href"]
                    #print(title)
                    #print(url)
                for watch_date in movie.select(".date"):
                    watch_date = watch_date.text
                    #print(watch_date)
                for intro in movie.select(".intro"):
                    intro = intro.text.strip()
                    #print(intro)
                    
                    myfilm_dict={
                        "title": title,
                        "watch_date":watch_date,
                        "intro":intro,
                        "url": url,
                    }
                    
                    myfilm.append(myfilm_dict)
                    
df = pd.DataFrame(myfilm)
df
```

# Hot tips:


* [0-9]{4}-[0-9]{2}-[0-9]{2} is a regex search pattern, meaning: 
* For example: Among 0-9 digits, I want the four digits first, then two digits plz, then another two digits, okay done!
* A RegEx (Regular Expression) is a sequence of characters that forms a search pattern. 
* RegEx can be used to check if a string contains the specified search pattern.
* My cheatsheet: https://stackoverflow.com/questions/64720582/python-pandas-split-datafram-column-into-two-new-column-and-drop-the-original-co


```python
df[["release_date","info"]] = df["intro"].str.extract(r'^([0-9]{4}-[0-9]{2}-[0-9]{2})(.*)',expand=False)
```

```python
df[["country","crew","crew2","3","4","5"]] = df["info"].str.split(")",expand=True)
```
